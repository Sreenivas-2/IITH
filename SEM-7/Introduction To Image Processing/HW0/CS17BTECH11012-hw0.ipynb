```python
# for plotting all the bitplanes of an image
def plotBitplanes(image, file):
    
    plane = []
    while (image).any():
        plane.append(image % 2)
        image = image // 2
        
    fig, ax = plt.subplots(nrows = 2, ncols = 4, figsize = (11, 11))
    
    ind = 0
    for row in ax:
        for col in row:
            col.imshow(plane[ind], cmap = 'gray')
            ind += 1
    
    fig.suptitle('Bitplane plot for ' + file, y = .9)
    plt.show()
```

```python
plotBitplanes(imgAirplane, 'Airplane - 7.1.02.tiff')
```

```python
plotBitplanes(imgAPC, 'APC - 7.1.08.tiff')
```

```python
# for modifying all the bitplanes of an image
def modifyBitplanes(image, file):
    
    plane = []
    ind = 0
    while ind < 8:
        image = image & ~(1 << ind)
        plane.append(image)
        ind += 1
        
    fig, ax = plt.subplots(nrows = 2, ncols = 4, figsize = (8, 8))
    
    ind = 0
    for row in ax:
        for col in row:
            col.imshow(plane[ind], cmap = 'gray')
            ind += 1
    
    fig.suptitle('Bitplane plot for ' + file, y = .9)
    
    plt.show()
```

```python
modifyBitplanes(imgAerial, 'Aerial - 5.2.09.tif')
```

```python
modifyBitplanes(imgAirplane, 'Airplane - 7.1.02.tiff')
```

```python
modifyBitplanes(imgAPC, 'APC - 7.1.08.tiff')
```

```python
# get histogram of the image
def binarizeImg(image, T):
    
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            if image[i][j] >= T:
                image[i][j] = 1
            else:
                image[i][j] = 0
                
    imgplot = plt.imshow(image, cmap = 'gray')
    return image
```

```python
def getConnectedComp(image):
    
    region = np.zeros(image.shape, dtype = np.int64)
    lookUp = {}
    k = 1
    
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            
            if image[i][j] == 0 and i == 0 and j == 0:
                region[i][j] = k
                lookUp[k] = k
                k += 1
                
            elif image[i][j] == 0 and i > 0 and j == 0:
                
                if image[i - 1][j] == 0:
                    region[i][j] = lookUp[region[i - 1][j]]
                else:
                    region[i][j] = k
                    lookUp[k] = k
                    k += 1
                    
            elif image[i][j] == 0 and i == 0 and j > 0:
                
                if image[i][j - 1] == 0:
                    region[i][j] = lookUp[region[i][j - 1]]
                else:
                    region[i][j] = k
                    lookUp[k] = k
                    k += 1
            
            elif image[i][j] == 0 and image[i - 1][j] == 1 and image[i][j - 1] == 1:
                region[i][j] = k
                lookUp[k] = k
                k += 1
                
            elif image[i][j] == 0 and image[i - 1][j] == 0 and image[i][j - 1] == 1:
                region[i][j] = lookUp[region[i - 1][j]]
                
            elif image[i][j] == 0 and image[i - 1][j] == 1 and image[i][j - 1] == 0:
                region[i][j] = lookUp[region[i][j - 1]]
                
            elif image[i][j] == 0 and image[i - 1][j] == 0 and image[i][j - 1] == 0:
                region[i][j] = lookUp[min(region[i - 1][j], region[i][j - 1])]
                
                if region[i - 1][j] != region[i][j - 1]:
                    lookUp[max(region[i - 1][j], region[i][j - 1])] = lookUp[region[i][j]]
                    
                    
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            
            if region[i][j] != 0:
                region[i][j] = lookUp[region[i][j]]
                
    return region
```

```python
def blobRemoval(image, region):
    
    dict_ = {}
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            
            if region[i][j] not in dict_:
                dict_[region[i][j]] = 1
            else:
                dict_[region[i][j]] += 1
            
    maxVal = 0
    maxEle = 0
    for key, value in dict_.items():
        if value > maxVal and key != 0:
            maxEle = key
            maxVal = value
            
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            
            if image[i][j] == 0 and region[i][j] != maxEle:
                image[i][j] = 1
                
    return image

def invert(image):
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            if image[i][j]:
                image[i][j] = 0
            else:
                image[i][j] = 1
```

```python
blobAirport = blobRemoval(binaryAirport, regionAirport)
invert(blobAirport)
region_blobAirport = getConnectedComp(blobAirport)
blobAirport = blobRemoval(blobAirport, region_blobAirport)
invert(blobAirport)
imgplot = plt.imshow(blobAirport, cmap = 'gray')
plt.suptitle('plot after minor blob removal - Airport - 5.3.02.tiff')
plt.show()
```

```python
def filter(image, i, j, s, n, case):
    
    if case == 'dilate':
        
        val = 0
        if s == 'cross':

            for row in range(-n, n + 1):
                if i - row >= 0 and i - row <= image.shape[0] - 1:
                    val = val or image[i - row][j]

            for col in range(-n, n + 1):
                if j - col >= 0 and j - col <= image.shape[1] - 1:
                    val = val or image[i][j - col]

        elif s == 'square':

            for row in range(-n, n + 1):
                for col in range(-n, n + 1):
                    if i - row >= 0 and i - row <= image.shape[0] - 1:
                        val = val or image[i - row][j]

                    if j - col >= 0 and j - col <= image.shape[1] - 1:
                        val = val or image[i][j - col]

        return val
    
    elif case == 'erode':
        
        val = 1
        if s == 'cross':

            for row in range(-n, n + 1):
                if i - row >= 0 and i - row <= image.shape[0] - 1:
                    val = val and image[i - row][j]

            for col in range(-n, n + 1):
                if j - col >= 0 and j - col <= image.shape[1] - 1:
                    val = val and image[i][j - col]

        elif s == 'square':

            for row in range(-n, n + 1):
                for col in range(-n, n + 1):
                    if i - row >= 0 and i - row <= image.shape[0] - 1:
                        val = val and image[i - row][j]

                    if j - col >= 0 and j - col <= image.shape[1] - 1:
                        val = val and image[i][j - col]

        return val
    
    elif case == 'median':
        
        val = 0
        if s == 'cross':

            for row in range(-n, n + 1):
                if i - row >= 0 and i - row <= image.shape[0] - 1:
                    val = max(val, image[i - row][j])

            for col in range(-n, n + 1):
                if j - col >= 0 and j - col <= image.shape[1] - 1:
                    val = max(val, image[i][j - col])

        elif s == 'square':

            for row in range(-n, n + 1):
                for col in range(-n, n + 1):
                    if i - row >= 0 and i - row <= image.shape[0] - 1:
                        val = max(val, image[i - row][j])

                    if j - col >= 0 and j - col <= image.shape[1] - 1:
                        val = max(val, image[i][j - col])

        return val
    
        

def function(image):
    
    J_CROSS_Dilate = np.zeros(image.shape)
    J_SQUARE_Dilate = np.zeros(image.shape)
    
    J_CROSS_Erode = np.zeros(image.shape)
    J_SQUARE_Erode = np.zeros(image.shape)
    
    J_CROSS_Median = np.zeros(image.shape)
    J_SQUARE_Median = np.zeros(image.shape)
    
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            J_CROSS_Dilate[i][j] = filter(image, i, j, 'cross', 2, 'dilate')
            J_SQUARE_Dilate[i][j] = filter(image, i, j, 'square', 1, 'dilate')
            
            J_CROSS_Erode[i][j] = filter(image, i, j, 'cross', 2, 'erode')
            J_SQUARE_Erode[i][j] = filter(image, i, j, 'square', 1, 'erode')
            
            J_CROSS_Median[i][j] = filter(image, i, j, 'cross', 2, 'median')
            J_SQUARE_Median[i][j] = filter(image, i, j, 'square', 1, 'median')
    
    fig, ax = plt.subplots(nrows = 1, ncols = 2, figsize = (8, 8))
    fig.subplots_adjust(hspace = .5)
    ax[0].imshow(J_CROSS_Dilate, cmap = 'gray')
    ax[1].imshow(J_SQUARE_Dilate, cmap = 'gray')
    fig.suptitle('plots for cross_Dilate (left) and square_Dilate(right)', y = .8)
    plt.show()
    
    print()
    fig, ax = plt.subplots(nrows = 1, ncols = 2, figsize = (8, 8))
    ax[0].imshow(J_CROSS_Erode, cmap = 'gray')
    ax[1].imshow(J_SQUARE_Erode, cmap = 'gray')
    fig.suptitle('plots for cross_Erode (left) and square_Erode(right)', y = .8)
    plt.show()
    
    print()
    fig, ax = plt.subplots(nrows = 1, ncols = 2, figsize = (8, 8))
    ax[0].imshow(J_CROSS_Median, cmap = 'gray')
    ax[1].imshow(J_SQUARE_Median, cmap = 'gray')
    fig.suptitle('plots for cross_Median (left) and square_Median(right)', y = .8)
    plt.show()
```

```python
def open_(image, case, n):
    
    JOpen = np.zeros(image.shape)
    
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            JOpen[i][j] = filter(image, i, j, case, n, 'erode')
            
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            JOpen[i][j] = filter(image, i, j, case, n, 'dilate')
    
    return JOpen
    

def close_(image, case, n):
    
    JClose = np.zeros(image.shape)
    
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            JClose[i][j] = filter(image, i, j, case, n, 'dilate')
            
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            JClose[i][j] = filter(image, i, j, case, n, 'erode')
    
    return JClose
```
